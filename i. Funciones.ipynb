# <font color='red'> <center>FUNCIONES</center></font>

Procesos reutilizables que se les puede dar un valor de entrada(parámetros) y pueden generar un valor de salida(retorno).

Para utilizar una función primero debe ser definida y luego debe ser llamada.

```python
import math
```

```python
#definiendo funcion:
def F(x): 
    operacion=x**2+math.sqrt(x)
    return operacion
```

Esta función se llama 'F', tiene como parámetro a 'x' y retorna 'operacion'.

```python
#llamando funcion:
r1=F(3)
print(r1)
r2=F(60)
print(r2)
```

## Parámetros: 

- Es lo que se le ingresa a la función. 
- Pueden ser de cualquier tipo de dato. 
- No es necesario que una función tenga parámetros y pueden ser más de uno.

```python
import random
```

```python
def generarNumeroConP(x):
    return random.randint(0,x)

def generarNumeroSinP():
    return random.randint(0,10)
```

```python
r1=generarNumeroConP(20)
print(r1)
r2=generarNumeroSinP() # No es necesario colocar ningun numero
print(r2)
```

## Retorno:

- Es lo que se devuelve o se genera de la función.
- Esto puede ser capturado en una variable externa a la función.
- Se puede retornar cualquier cantidad y tipo de datos.
- Una función puede llevar o no un retorno.

```python
def sumaSinR(x,y):
    suma=x+y
    print("El resultado de la suma de {} con {} es igual a {}".format(x,y,suma))

def sumaConR(x,y):
    suma=x+y
    print("El resultado de la suma de {} con {} es igual a {}".format(x,y,suma))
    return suma
```

```python
sumaSinR(6,7) # cuando no se retorna, no es necesario capturar o almacenar un resultado
resultado=sumaConR(3,2)
print(resultado)
```

```python
# Retorno múltiple en una función
def operaciones(x,y):
    suma=x+y
    resta=x-y
    mult=x*y
    return suma,resta,mult

a = int(input("Ingrese n1: "))
b = int(input("Ingrese n2: "))
s,r,m=operaciones(a,b)
print(s)
print(r)
print(m)
```

```python
def funcion(variable):
  if variable == 0:
    return None
  else:
    return variable+3

x = funcion(0)
print(x)
```
