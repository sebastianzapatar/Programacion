# <font color='red'> <center>VARIABLES</center></font>

Una variable guarda un dato en la memoria RAM del computador. Estas se definen por un nombre y por un tipo. 

Tipos de variables en Python:
- Numéricas (Enteros y reales)
- Cadena (Texto) 
- Binarias (Booleanas)
- Colecciones (Listas, diccionarios, tuplas, conjuntos, entre otros)

Python a diferencia de otros lenguajes no requiere definir el tipo de la variable, si no que dependiendo del valor con el que se inicializa se define el tipo.

### Variables numéricas:
- **Entero (int)**: Que no tienen una parte decimal.
- **Real (float)**: Números que tienen una parte decimal después de un punto.


```python
a=3 #Variable de  tipo entero
b=5.54555654  #Variable de tipo real o flotante
```

```python
print(a)
print(b)
```

Para validar que a cada variable le corresponda el tipo de dato correcto, se puede hacer lo siguiente:

```python
print(type(a))
print(type(b))
```

### Variables de texto o cadena (str):
Son aquellas formadas por un carácter o un conjunto de caracteres como lo pueden ser letras, números o caracteres especiales (? , .  [] {} etc).
Siempre el valor de dicha variable se encuentra entre comillas " ".

```python
nombre = "HOLA"
codigo = "a234asd35"
```

```python
print(nombre) 
print(codigo)  
```

```python
print(type(nombre)) 
print(type(codigo)) 
```
