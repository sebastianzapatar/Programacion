## Aritméticos:

Si se aplican a variables numéricas realizan las operaciones matemáticas básicas:

```python
a=2
b=3.5
c=a*b # *: multiplicar /: dividir -: restar +:sumar
print(c)
# Operaciones que tengan al menos un flotante, genera un resultado flotante
print(type(c))
```

```python
# Operaciones entero con entero, dan como resultado un entero a excepción de la división
d=4
e=8
f=e/d
print(f)
print(type(f))
```

```python
#Para elevar un numero a una potencia se utiliza **
print(d**2)
```

## Relacionales:
 Sirven para comparar dos valores y generan como resultado:
 - True, si la comparación es verdadera.
 - False, si la comparación es falsa.
 
Los resultados se pueden guardar en variables llamadas booleanas o bool.

```python
variableBool=False
print(variableBool)
print(type(variableBool))
```

- Mayor/menor que: <,>
- Mayor/menor e igual que: <=,>=
- Igual que: ==
- Diferente que: !=

```python
n1=3
n2=3
comparacion=n1!=n2
print(comparacion)
print(type(comparacion))
```

```python
t1 = "hola"
print(t1)
t2 = "HOLA".lower()
print(t2)
print(t1==t2)
```

```python
texto = input("Hola ingrese texto ").lower()
print(texto)
```

## Lógicos:
Evalúa dos variables o datos de tipo bool, dando como resultado otro bool.
 
 - Conjunción(and): Devuelve True cuando los dos valores son True, para todo caso contrario, devuelve False.
 - Disyunción(or): Devuelve False cuando los dos valores son False, para todo caso contrario, devuelve True.
 - Negación(not): Devuelve True si el valor es False y devuelve False si el valor es True.

```python
verdadero = True
falso = False

conjuncion=verdadero and falso
print(conjuncion)

disyuncion=falso or falso
print(disyuncion)

negacion=not verdadero
print(negacion)
```

## Expresiones compuestas:
Involucran más de un tipo de operador y se pueden solucionar empleando las reglas de precedencia:
1. Las expresiones encerradas en paréntesis.
2. Las expresiones aritméticas por sus propias reglas.
3. Las expresiones relacionales.
4. Las expresiones lógicas(not,and,or).

```python
a = 10
b = 5
resultado = a*b-2**b >= 20 and not (a/(b+1)) != 2
print(resultado)
```
