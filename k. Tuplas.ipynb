# <font color='red'> <center>TUPLAS</center></font>
Es una colección de datos inmutable que organiza sus elementos según un índice. Sus elementos se separan por comas y se encierran entre paréntesis.

```python
tupla = (2,"texto",[10,23,35],-4.5)
print(tupla)
print(type(tupla))
```

```python
print(tupla[-2])
print(tupla[1])
print(tupla[-2][-3])
```

Al ser inmutable no se le puede agregar ni editar elementos:

```python
tupla[2]=3
```

```python
tupla.append(10)
```

Todas las tuplas se pueden convertir en listas en caso de tener que agregar o editar algun elemento para luego ser convertidas otra vez en tuplas:

```python
lista=list(tupla)
lista[2]=3
lista.append(10)
tupla=tuple(lista)
print(tupla)
print(lista)
```

Es común que el Python nos devuelva información en forma de tupla:

```python
def funcion():
    numero=3
    texto='Hola! Que tal?'
    return numero,texto
```

```python
datos = funcion()
print(datos)
print(type(datos))
```

```python
numero = datos[0]
texto = datos[-1]
print(numero)
print(texto)
```
