# Lista:

Es una colección de datos mutable que organiza sus elementos según un índice. Sus elementos se separan por comas y se encierran entre corchetes.

**Mutable:** Que se pueden modificar sus elementos. Se pueden agregar o eliminar elementos.

**Índice:** Es un número entero que identifica un elemento almacenado en la lista.

Para crear una lista:

Cree una lista de N números aleatorios:

```python
import random
```

```python
print(random.randint(1,6))
print(round(random.uniform(0,10),3))
```

```python
N=int(input('Ingrese cantidad de numeros: '))
lista=[]
print(lista)
for n in range(N):
    lista.append(random.randint(0,10))
    print(lista)
print(lista)
```

Cree una lista de N números aleatorios utilizando el concepto de comprensión de listas

```python
N=int(input('Ingrese cantidad de numeros: '))
lista = [random.randint(0,10) for n in range(N)]
print(lista)
# Hace lo mismo que el programa anterior solo que se implemente en una sola linea de codigo
```

Realice la suma total de números almacenados en una lista:

```python
lista = [random.randint(0,5) for x in range(5)]
print(lista)
```

```python
N=len(lista)
suma=0
for i in range(N):
  # print(lista[i])
  suma+=lista[i] # suma = suma + lista[i]
print(suma)
```

```python
suma=sum(lista)
print(suma)
```

Otras funciones útiles:

```python
lista = [random.randint(0,5) for x in range(5)]
print(lista)
minimo = min(lista)
print('Valor minimo:',minimo)
maximo = max(lista)
print('Valor maximo:',maximo)
copiaLista = lista.copy()
print('Lista copia:',copiaLista)
lista.sort(reverse=False) # True: desc - False:asc
print('Lista ordenada:',lista)
```
