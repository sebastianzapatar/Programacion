Le permiten a un programa que tenga más de un posible camino.

La sentencia **if** permite evaluar una condición y si esta es ***verdadera*** se ejecuta un proceso específico.

```python
variable = -3
if variable > 0: # Siempre que se pongan : la siguiente linea de codigo tiene que estar indentada(corrida hacia la derecha)
  print('Numero positivo')
print('Termina programa')
```

Se pueden utilizar expresiones compuestas como condiciones:

```python
a=-3
b=-5
if a>0 or b>0:
    print('Uno de los dos numeros es positivo')
print("Termina el programa")
```

Condicional compuesto o anidado:

```python
c=0
d=100
if c==0:
    print('C es igual a cero')
    if d<100:
        print('D es menor que 100')
    print('Finaliza programa')
```

 La sentencia **else** se acopla a un if para comprobar el caso contrario (en el que no se cumple la condición).

```python
n = 37
if n % 2 == 0:
    print(n,"es un número par")
else:
    print(n,"es un número impar")
```

La sentencia **elif** se acopla a un if u otro elif para comprobar múltiples condiciones, siempre que las anteriores no se ejecuten.

```python
nota = float(input("Introduzca una nota: "))
if nota>5 or nota<0:
    print("Nota no valida")
elif nota >= 4.5:
    print("Excelente")
elif nota >= 3.8:
    print("Sobresaliente")
elif nota >= 3: 
    print("Aceptable")
elif nota >= 1.5:
    print("Insuficiente")
else:
    print("Deficiente")    
```
