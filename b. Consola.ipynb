##### La consola es la herramienta por defecto que nos permite interactuar con nuestro código ya sea a partir de una entrada o una salida.
### **SALIDA:**

Las salidas por consola se generan a partir de la función print(), siendo esta capaz de mostrar el estado de una variable o generar un mensaje en específico.

```python
nombre = "Cristiano Ronaldo"
saludo = f"Hola, soy {nombre}"
print(saludo)
```

```python
x = 10.5
y = 2
mensaje = f"{x}+{y}={x+y}"
print(mensaje)
```

### **ENTRADA:**

Para generar una entrada mediante la consola se utiliza la función input(), siendo esta una herramienta que permite darle un valor inicial a una variable de manera dinámica. Hasta que no se indique cual es la entrada, el programa se detiene.

```python
entrada = input("Ingrese un valor: ")
print(entrada)
print(type(entrada))
```

Por defecto las variables que se generan a partir de input() son de tipo str.
 Si se desea convertir esta variable a int o float se utilizan las funciones int() o float():

```python
oe = input("Ingresar número entero: ")
print(type(oe))
oe = int(oe)
print(type(oe))
print(oe+610)
```

O de manera más eficiente (pero utilizando float):

```python
entradaFloat = float(input("Ingresar número flotante: "))
print(type(entradaFloat))
print(entradaFloat)
```
