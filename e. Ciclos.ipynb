¿Qué pasa si se necesitan ahora los números hasta el 234432432423?

Claramente la solución anterior no es viable para esta nueva situación.

Cuando se tenga una serie procesos repetitivos y que tenga cierta similitud o tendencia entre línea y línea de código, lo correcto es utilizar alguna de las sentencias disponibles para generar ciclos.

Tipos de sentencias:

- for: cuando sabemos la cantidad de repeticiones
- while: cuando no sabemos la cantidad de repeticiones

## for:

```python
N=5
for x in range(N):
    print(x)
```

```python
M=0
N=5
for x in range(M,N):
    print(x)
```

```python
M=-10
N=5
S=3
for x in range(M,N+1,S):
    print(x)
```

Genere un programa que sea capaz de identificar si los numeros desde M hasta N son pares o impares:

```python
M=2
N=154
for x in range(M,N):
  # print(x)
    if x%2==0:
        print('Numero par:',x)
    else:
        print('Numero impar:',x) 
```

## while:

Tenga en cuenta que esta sentencia siempre depende de una condición.

Replique el funcionamiento de las 3 primeras casillas de la sentencia for pero utilizando ahora while:

```python
M=2
N=20
S=2
while M<N+1:
    print(M)
    M+=S # M=M+S
print('Fin')
```

Cree un programa que simule la recolección de un dinero.

```python
meta=100
acumulado=0
while acumulado<meta:
    recolectar=float(input('Ingrese valor: '))
    acumulado+=recolectar  # acumulado=acumulado+recolectar
    print('Acumulado actual:',acumulado)
```

Cree un menú por consola para sumar una pareja de números. El menú debe contar con las opciones de 'salir' y de 'sumar'. En caso de que no se ingrese una opción adecuada se debe mostrar por consola un mensaje que lo indique.

```python
while True: #Ciclo infinito
    opcion = input('Ingrese opcion: ').upper()
    print(opcion)
    if opcion == 'SUMAR':
        n1=float(input('Ingrese numero uno a sumar: '))
        n2=float(input('Ingrese numero dos a sumar: '))
        resultado=n1+n2
        print("El resultado de la suma es",resultado)
    elif opcion == 'salir':
        break # Cuando se ejecuta este comando, el ciclo(sea for o while) se rompe/se detiene.
    else:
        print('Opcion no valida')
print('Fin')
```
